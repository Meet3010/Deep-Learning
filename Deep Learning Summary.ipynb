##### What is the Activation Function:
If we do not apply a Activation function then the output signal would simply be a simple linear function.A linear function is just a polynomial of one degree. Now, a linear equation is easy to solve but they are limited in their complexity and have less power to learn complex functional mappings from data. A Neural Network without Activation function would simply be a Linear regression Model, which has limited power and does not performs good most of the times

Activation Function, also known as transfer function, introduces non-linearity in neural networks.
This non-linear transformation is introduced to learn the complex underlying patterns in the data.
Without this function, a neuron simply resembles a linear regression.

* Sigmoid
* Tanh
* ReLu
* LeakyReLu
* Softmax

#### Activation Function

Activation Function, also known as transfer function, introduces non-linearity in neural networks.
This non-linear transformation is introduced to learn the complex underlying patterns in the data.
Without this function, a neuron simply resembles a linear regression.

1. Sigmoid Function:

It scales all the values between 0 to 1, thus used for predicting probalities.
Here problem of varnishing gradient arises,which makes it unsuitable for propogation.

2. Hyperbolic Tangent(tanh):

It scales all the values between -1 to 1.
This function solves the "varnishing gradients" problem to some extent.

3. RELU(Rectified Linear Unit):

This is the most useful and relatively simple function, max(0,z).
ReLU minimises the varnishing gradients problem to a great degree.
i.e from 0 to positive (max value of Z)

Softmax Function:-

It is a generalization of a sigmoid function.
Softmax ouptuts values between 0 and 1 such that their sum is 1.
Thus, Softmax is applied to the final layer of a network in a multi-class classification.

#### Cost Function:

A function called Cost Function or Loss Function tells how well a neural network performs.
The difference between the actual value and the predicted value is called the Cost Fucntion.
Its main aim is to minimize the cost to get the best performing neural network.
Its main drawback is that if we apply cost function , it will slow the learning rate of the neural networks

So we use another instead of this called Cross Entropy.

Cross Entropy:
It reduces the cost function, also it fastens the learning rate of the cost fucntion

#### Gradient Descent

A neural network predicts the output using Forward Propogation
Forward Propogation starts with the random weights.
The output of forward propogation is compared against actual ouput to compute the cost of the network.
To mitigate the cost,changes must be made to the weight values.
This is achieved using Gradient Descent.

Gradient Descent helps the neural network learn the optimal value of weights to generate the correct output and minimize the cost.

#### Back Propagation

An algorithm to improve the accuracy of prediction and quick calculation of derivates.

Backpropogation is used to update the weights from output to input.

The process of revising the weights after the checking the output and re doing the same process again from the input is called Back Propagation method. It helps in reducing the cost function

Learning Rate(Î±):

A hyperparameter that controls how much to update a network each time weights are updated to minimize the calculated cost.

#### Keras:

High level API that allows to build train, evaluate and excecute all types of neural networks.

```python
# Importing the model and layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
# we will input the model into a variable
model=Sequential()
```

```python
# Define the layers of the model
# Watch class 1 of the program at 3:05 mins to understand the dense and sequential
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

It is an ideal practice to choose the no of neurons to the power of 2 like 16,32,64,128,256,512. we can use the other nos also.

The more the no of neurons the more will be the time taken to excecute the model.

NOTE:

Since its a multiclassification problem have more than 2 outputs we are using the 'softmax' activation function, if it would a binary classification of 0 and 1(two values) we should use 'sigmoid' activation function


```python
# We need to compile the model
model.compile(optimizer='sgd',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Checking by plotting the same
plt.subplots(figsize = (10,8))
plt.title('Counts in numbers to their labels ')
sns.countplot(x=y, data=train)
plt.show()
```

```python
# so reshaping the data
X_train=X_train.reshape(X_train.shape[0],28,28,1)
X_val=X_val.reshape(X_val.shape[0],28,28,1)
test=test.reshape(test.shape[0],28,28,1)

# checking the shape of the data
X_train.shape,X_val.shape,test.shape
```
